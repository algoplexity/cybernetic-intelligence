# Prerequisites

1. Install Ollama from https://ollama.com/
2. Pull the model you want to use (e.g., `ollama pull llama2` or `ollama pull mistral`)
3. Ensure Ollama is running locally (default port: 11434)

```python
# =======================================================================================
#  The Official and Correct Colab Setup for Ollama
# =======================================================================================
#
# This script uses the official Ollama installer, which automatically handles
# finding the latest version, downloading it, and placing it in the system path.
# This avoids all previous errors related to incorrect download links.
#
import os
import asyncio

# --- Configuration ---
MODEL_TO_PULL = "gpt-oss:20b"

async def run_shell_command(command):
    """Asynchronously runs a shell command and streams its output."""
    process = await asyncio.create_subprocess_shell(
        command,
        stdout=asyncio.subprocess.PIPE,
        stderr=asyncio.subprocess.PIPE
    )

    # Stream stdout and stderr
    async def read_stream(stream, prefix):
        while True:
            line = await stream.readline()
            if line:
                print(f"{prefix}: {line.decode().strip()}")
            else:
                break

    await asyncio.gather(
        read_stream(process.stdout, "stdout"),
        read_stream(process.stderr, "stderr")
    )

    return await process.wait()

async def setup_ollama():
    """Main function to setup and configure Ollama."""

    # --- Step 1: Install Ollama using the Official Installer Script ---
    print("--- Step 1: Installing Ollama using the official script... ---")
    install_command = "curl -fsSL https://ollama.com/install.sh | sh"
    await run_shell_command(install_command)
    print("\n‚úÖ Ollama installation script finished.")

    # --- Step 2: Start the Ollama Server in the Background ---
    # The official installer places ollama in the system path.
    # We use nohup to ensure the server keeps running in the background.
    print("\n--- Step 2: Starting Ollama server in the background... ---")
    # Start the server and detach it from the current process
    subprocess.Popen(["ollama", "serve"], start_new_session=True, stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
    print("Waiting for server to initialize (10 seconds)...")
    await asyncio.sleep(10)
    print("‚úÖ Server should be running.")

    # --- Step 3: Pull the Desired LLM ---
    print(f"\n--- Step 3: Pulling model '{MODEL_TO_PULL}' ---")
    print("This may take several minutes...")
    pull_command = f"ollama pull {MODEL_TO_PULL}"
    await run_shell_command(pull_command)

    # --- Final Check ---
    print("\n--- Final Check: Listing installed models ---")
    await run_shell_command("ollama list")

    print("\n\n‚úÖ‚úÖ‚úÖ SETUP COMPLETE. You can now run the rest of the notebook. ‚úÖ‚úÖ‚úÖ")


# Run the asynchronous setup function
await setup_ollama()
```

```python
import subprocess
import time
import requests
import os

def is_ollama_running():
    """Check if Ollama is running"""
    try:
        response = requests.get("http://localhost:11434/api/version", timeout=2)
        return response.status_code == 200
    except:
        return False

if not is_ollama_running():
    print("Starting Ollama server...")
    # Start Ollama in the background using subprocess.Popen
    # This won't block the notebook
    process = subprocess.Popen(
        ['ollama', 'serve'],
        stdout=subprocess.DEVNULL,
        stderr=subprocess.DEVNULL,
        start_new_session=True  # This detaches the process
    )

    # Wait a few seconds for Ollama to start
    for i in range(10):
        time.sleep(1)
        if is_ollama_running():
            print("‚úÖ Ollama server started successfully!")
            break
    else:
        print("‚ö†Ô∏è Ollama might take longer to start. Please wait and re-run the next cell.")
else:
    print("‚úÖ Ollama is already running!")
```

# üòá Basic Usage with AdalFlow Generator

Let's start with the simplest way to use Ollama with AdalFlow's Generator component. We'll use the `gpt-oss:20b` model for these examples.

## Simple Text Generation

The Generator is AdalFlow's main component for interacting with language models. Here's how to use it with Ollama:

```python
from adalflow.components.model_client.ollama_client import OllamaClient
from adalflow.core import Generator

# Initialize the Generator with OllamaClient
# Using gpt-oss:20b model as shown in the test file
generator = Generator(
    model_client=OllamaClient(host="http://localhost:11434"),
    model_kwargs={
        "model": "gpt-oss:20b",  # Using gpt-oss model
    }
)

# Test with a simple prompt
response = generator.call(prompt_kwargs={"input_str": "Hello! What are the benefits of using local LLMs"})
print("Response:")
print(response.data)
```

## Asynchronous Call

For better performance, you can use the async version with `acall`:

```python

# Using async call with acall
output = await generator.acall(prompt_kwargs={"input_str": "What are the advantages of async programming?"})

print("Async Response:")
print(output.data)
```

## Streaming Responses

For real-time output, you can stream responses directly from Ollama:

```python
# pull qwen2:0.5b

!ollama pull qwen2:0.5b
```

```python
from adalflow.components.model_client.ollama_client import OllamaClient
from adalflow.core import Generator

stream_generator = Generator(
    model_client=OllamaClient(host="http://localhost:11434"),
    model_kwargs={
        "model": "gpt-oss:20b",
        "stream": True,  # Enable streaming
    }
)

# async call with streaming
output = await stream_generator.acall(prompt_kwargs={"input_str": "Why is the sky blue?"})

async for chunk in output.raw_response:
    print(chunk["message"]["content"], end='', flush=True)

print(output)
```

```python
# sync Call with streaming
output = stream_generator.call(prompt_kwargs={"input_str": "Why is the sky blue?"})

for chunk in output.raw_response:
    print(chunk["message"]["content"], end='', flush=True)

print(output)
```
